```python
import csv

sales = 0
costs = 0

with open("sales.csv", "r") as f:
    reader = csv.DictReader(f, delimiter=",")
    for row in reader:
            sales += int(row['Sales'])
            costs += int(row['Costs'])
            
print ("Sales: ", sales)
print ("Costs: ", costs)
```
